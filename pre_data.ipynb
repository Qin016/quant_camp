```python
import numpy as np
import pandas as pd
import os
from pathlib import Path

# 定义数据路径
data_dir = Path("./future_L2/train/20220722")
file_name = "IC_2208_L1_M.parquet"  # IC主力合约
file_path = data_dir / file_name

# 检查文件是否存在
if file_path.exists():
    print(f"正在读取文件: {file_path}")
    
    # 使用pandas读取parquet文件
    df = pd.read_parquet(file_path)
    
    # 显示基本信息
    print(f"\n数据形状: {df.shape}")
    print(f"列名: {list(df.columns)}")
    print(f"数据类型:")
    print(df.dtypes)
    
    # 显示前几行数据
    print(f"\n前5行数据:")
    print(df.head())
    
    # 显示统计信息
    print(f"\n数值列统计信息:")
    print(df.describe())
    
    # 检查时间列
    if 'TradingTime' in df.columns:
        print(f"\n时间范围:")
        print(f"开始时间: {df['TradingTime'].min()}")
        print(f"结束时间: {df['TradingTime'].max()}")
        print(f"数据频率: {len(df)} 条记录")
    
else:
    print(f"文件不存在: {file_path}")
    print("可用文件列表:")
    for f in data_dir.glob("*.parquet"):
        print(f"  {f.name}")   
```
